```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
dataset = sc_X.fit_transform(dataset)
dataset=pd.DataFrame(dataset, columns=["Gender","Age","Annual Salary","Credit Card Debt","Net Worth","Car Purchase Amount"])
```

```python
# visualize the relationship between the features and the response using scatterplots
sb.pairplot(dataset, x_vars=["Gender","Age","Annual Salary","Credit Card Debt","Net Worth"], y_vars="Car Purchase Amount", height=7, aspect=0.7)
```

```python
X = dataset.iloc[:,:-1].values.reshape(500,5)
X.shape
# iloc -> pandas dataframe to pandas series
# for reshaping pandas series .values.reshape()
# good practice to use reshape
```

```python
y = dataset.iloc[:,-1].values.reshape(500,1)
y.shape
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
# coefficients
coeff_df = pd.DataFrame(regressor.coef_.squeeze(), dataset.columns[:-1].values, columns=['Coefficient'])  
coeff_df
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('Coefficient of determination: %.2f' % r2_score(y_test, y_pred))
```
