```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Car_Purchasing_Data.csv',encoding='latin1')
```

```python
dataset = dataset.drop(columns = ['Customer Name', 'Customer e-mail', 'Country'])
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum(axis = 0) # dataset.isnull().any()
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
dataset = sc_X.fit_transform(dataset)
dataset=pd.DataFrame(dataset, columns=["Gender","Age","Annual Salary","Credit Card Debt","Net Worth","Car Purchase Amount"])
```

```python
import seaborn as sb
%matplotlib inline

# visualize the relationship between the features and the response using scatterplots
sb.pairplot(dataset, x_vars=["Gender","Age","Annual Salary","Credit Card Debt","Net Worth"], y_vars="Car Purchase Amount", height=7, aspect=0.7)
```

```python
X = dataset.iloc[:,:-1].values.reshape(500,5)
X.shape
# iloc -> pandas dataframe to pandas series
# for reshaping pandas series .values.reshape()
# good practice to use reshape
```

```python
y = dataset.iloc[:,-1].values.reshape(500,1)
y.shape
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
import tensorflow as tf

# Initializing the ANN
model = tf.keras.models.Sequential()

# Adding the input layer and the first hidden layer
model.add(tf.keras.layers.Dense(units=10, input_dim=5, kernel_initializer='normal', activation='relu'))

# Adding the second hidden layer
model.add(tf.keras.layers.Dense(units=5, activation='relu'))

# Adding the output layer
model.add(tf.keras.layers.Dense(units=1, activation='linear'))

# Summary
model.summary()

# Compiling the ANN
model.compile(optimizer = 'adam', loss='mse', metrics=['mse','mae'])
```

```python
# Training the ANN on the Training set
model.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('Coefficient of determination: %.2f' % r2_score(y_test, y_pred))
```
